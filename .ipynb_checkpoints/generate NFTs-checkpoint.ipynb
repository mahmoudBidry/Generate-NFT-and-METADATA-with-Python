```python
from PIL import Image 
import random
import json
import os
```

```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%

background = ["Name_background1","Name_background2","Name_background3","Name_background4","Name_background5","Name_background6",
              "Name_background7","Name_background8","Name_background9","Name_background10","Name_background11","Name_background12",
              "Name_background13","Name_background14"] 
background_weights = [8, 8, 8, 8, 8, 8, 8, 8, 8, 6, 6, 6, 6, 4]

beard_and_mustache = ["Name_beard_and_mustache1", "Name_beard_and_mustache2", "Name_beard_and_mustache3", "Name_beard_and_mustache4",
                      "Name_beard_and_mustache5", "Name_beard_and_mustache6", "Name_beard_and_mustache7", "Name_beard_and_mustache8",
                      "Name_beard_and_mustache9"]
beard_and_mustache_weights = [15, 15, 15, 15, 15, 15, 7, 2, 1]

body = ["Name_body1", "Name_body2", "Name_body3", "Name_body4", "Name_body5"]
body_weights = [10, 25, 25 , 25 , 15]

eyes = ["Name_eyes1", "Name_eyes2", "Name_eyes3", "Name_eyes4", "Name_eyes5"]
eyes_weights = [10, 25, 25 , 25 , 15]

glasses = ["Name_glasses1", "Name_glasses2", "Name_glasses3", "Name_glasses4", "Name_glasses5"]
glasses_weights = [10, 25, 25 , 25 , 15]

mouth = ["Name_mouth1", "Name_mouth2"]
mouth_weights = [70, 30]

outfits = ["Name_outfits1", "Name_outfits2", "Name_outfits3", "Name_outfits4", "Name_outfits5",
           "Name_outfits6", "Name_outfits7"]
outfits_weights = [15, 15, 15, 15, 15, 15, 10]
```

```python
#Classify traits

background_files = {
    "Name_background1": "background1",
    "Name_background2": "background2",
    "Name_background3": "background3",
    "Name_background4": "background4",
    "Name_background5": "background5",
    "Name_background6": "background6",
    "Name_background7": "background7",
    "Name_background8": "background8",
    "Name_background9": "background9",
    "Name_background10": "background10",
    "Name_background11": "background11",
    "Name_background12": "background12",
    "Name_background13": "background13",
    "Name_background14": "background14"
}

beard_and_mustache_files = {
    "Name_beard_and_mustache1": "beard and mustache1",
    "Name_beard_and_mustache2": "beard and mustache2",
    "Name_beard_and_mustache3": "beard and mustache3",
    "Name_beard_and_mustache4": "beard and mustache4",
    "Name_beard_and_mustache5": "beard and mustache5",
    "Name_beard_and_mustache6": "beard and mustache6",
    "Name_beard_and_mustache7": "beard and mustache7",
    "Name_beard_and_mustache8": "beard and mustache8",
    "Name_beard_and_mustache9": "beard and mustache9"

}

body_files = {
    "Name_body1": "body1",
    "Name_body2": "body2",
    "Name_body3": "body3",
    "Name_body4": "body4",
    "Name_body5": "body5"
}

eyes_files = {
    "Name_eyes1": "eyes1",
    "Name_eyes2": "eyes2",
    "Name_eyes3": "eyes3",
    "Name_eyes4": "eyes4",
    "Name_eyes5": "eyes5"     
}

glasses_files = {
    "Name_glasses1": "glasses1",
    "Name_glasses2": "glasses2",
    "Name_glasses3": "glasses3",
    "Name_glasses4": "glasses4",
    "Name_glasses5": "glasses5"     
}

mouth_files = {
    "Name_mouth1": "mouth1",
    "Name_mouth2": "mouth2"
}

outfits_files = {
    "Name_outfits1": "outfits1",
    "Name_outfits2": "outfits2",
    "Name_outfits3": "outfits3",
    "Name_outfits4": "outfits4",
    "Name_outfits5": "outfits5",
    "Name_outfits6": "outfits6",
    "Name_outfits7": "outfits7"
}
```

```python
## Generate Traits

TOTAL_IMAGES = 100 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["background"] = random.choices(background, background_weights)[0]
    new_image ["beard_and_mustache"] = random.choices(beard_and_mustache, beard_and_mustache_weights)[0]
    new_image ["body"] = random.choices(body, body_weights)[0]
    new_image ["eyes"] = random.choices(eyes, eyes_weights)[0]
    new_image ["glasses"] = random.choices(glasses, glasses_weights)[0]
    new_image ["mouth"] = random.choices(mouth, mouth_weights)[0]
    new_image ["outfits"] = random.choices(outfits, outfits_weights)[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

#print("Are all images unique?", all_images_unique(all_images))
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1

#print(all_images)
```

```python
# Get Trait Counts

background_count = {}
for item in background:
    background_count[item] = 0
    
beard_and_mustache_count = {}
for item in beard_and_mustache:
    beard_and_mustache_count[item] = 0

body_count = {}
for item in body:
    body_count[item] = 0
    
eyes_count = {}
for item in eyes:
    eyes_count[item] = 0
    
glasses_count = {}
for item in glasses:
    glasses_count[item] = 0
    
mouth_count = {}
for item in mouth:
    mouth_count[item] = 0
    
outfits_count = {}
for item in outfits:
    outfits_count[item] = 0

for image in all_images:
    background_count[image["background"]] += 1
    beard_and_mustache_count[image["beard_and_mustache"]] += 1
    body_count[image["body"]] += 1
    eyes_count[image["eyes"]] += 1
    glasses_count[image["glasses"]] += 1
    mouth_count[image["mouth"]] += 1
    outfits_count[image["outfits"]] += 1
    
print(background_count)
print(beard_and_mustache_count)
print(body_count)
print(eyes_count)
print(glasses_count)
print(mouth_count)
print(outfits_count)
```

```python
### Generate Images

os.mkdir(f'./images')

for item in all_images:

    im1 = Image.open(f'./NFT parts/background/{background_files[item["background"]]}.png').convert('RGBA')    
    im2 = Image.open(f'./NFT parts/beard and mustache/{beard_and_mustache_files[item["beard_and_mustache"]]}.png').convert('RGBA')
    im3 = Image.open(f'./NFT parts/body/{body_files[item["body"]]}.png').convert('RGBA')
    im4 = Image.open(f'./NFT parts/eyes/{eyes_files[item["eyes"]]}.png').convert('RGBA')
    im5 = Image.open(f'./NFT parts/glasses/{glasses_files[item["glasses"]]}.png').convert('RGBA')
    im6 = Image.open(f'./NFT parts/mouth/{mouth_files[item["mouth"]]}.png').convert('RGBA')
    im7 = Image.open(f'./NFT parts/outfits/{outfits_files[item["outfits"]]}.png').convert('RGBA') 

    #Create each composite
    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)
    com5 = Image.alpha_composite(com4, im6)
    com6 = Image.alpha_composite(com5, im7)

                     

    #Convert to RGB
    rgb_im = com6.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)
    
```

```python

#### Generate Metadata for all Traits 
os.mkdir(f'./metadata')

METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)

# Changes this IMAGES_BASE_URL to yours 
IMAGES_BASE_URL = "https://gateway.pinata.cloud/ipfs/QmShRfUitXELLoz1ZcJWbsVxcnZrkmNapUVLWBwp18KaXh/"
PROJECT_NAME = "NFT_CREATOR"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URL + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Face", i["Face"]))
    token["attributes"].append(getAttribute("Ears", i["Ears"]))
    token["attributes"].append(getAttribute("Eyes", i["Eyes"]))
    token["attributes"].append(getAttribute("Hair", i["Hair"]))
    token["attributes"].append(getAttribute("Mouth", i["Mouth"]))
    token["attributes"].append(getAttribute("Nose", i["Nose"]))

    with open('./metadata/' + str(token_id) + ".json", 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
